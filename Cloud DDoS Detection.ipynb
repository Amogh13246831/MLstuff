```python
# cloud computing paper: training and testing of DDoS attack detector
import random
import numpy as np
import pandas as pd

import keras as kr
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, TimeDistributed
```

```python
#training data
intervals = (1, 0.5, 0.1, 0.01, 0.005, 0.001)
no_att_int = (1, 0.5, 0.1, 0.01)

ddos_train = []
for _ in range(4):
    with_atk = random.choices(intervals, k=1000)
    #without_atk = random.choices(intervals, k=300)
    #with_atk.extend(without_atk)
    ddos_train.append(with_atk)
 
tr = pd.DataFrame(ddos_train)
train_df = pd.DataFrame(tr.T.values, columns = ("VM1", "VM2", "VM3", "VM4"))
print(train_df.shape)
```

```python
target = [0] * train_df.shape[0]
vals = train_df.values 

for i in range(train_df.shape[0]):
    if any(vals[i] == 0.001):
        target[i] = 1

print(target[:10])
train_df.head(10)
```

```python
X = train_df.values
Y = target

model = Sequential()
model.add(Dense(5, input_dim=4, activation='tanh'))
model.add(Dense(1, activation='tanh'))

model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
history = model.fit(X, Y, epochs=100, batch_size=10)
```

```python
loss, accuracy = model.evaluate(X, Y)
print("\nLoss: %.2f, Accuracy: %.2f%%" % (loss, accuracy*100))
```

```python
x_train = train_df.values.reshape(1000,1,4)
y_train = pd.DataFrame(target).values.reshape(1000, 1,1)

print(x_train[:10], y_train[:10], sep="\n")
model=Sequential()
model.add(SimpleRNN(input_dim=4, output_dim=5, return_sequences = True))
model.add(TimeDistributed(Dense(output_dim = 1, activation  =  "relu")))
model.compile(loss = "mse", optimizer = "adam", metrics=["accuracy"])
model.fit(x_train, y_train, nb_epoch = 200, batch_size = 32)
```
